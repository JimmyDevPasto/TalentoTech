# Tarea: consultar qué es y cómo funciona la matriz de confusión (clasificación)

La matriz de confusión es una herramienta fundamental en la evaluación de algoritmos de clasificación en aprendizaje automático y estadística. Esta matriz permite visualizar el desempeño de un modelo clasificador al comparar las predicciones realizadas con las clases reales de los datos.

Estructura de la Matriz de Confusión
Supongamos que estamos trabajando con un problema de clasificación binaria (dos clases: positiva y negativa). La matriz de confusión para este caso tiene la siguiente estructura:

Clase Real Positiva (P)	Clase Real Negativa (N)
Predicción Positiva (P')	Verdaderos Positivos (VP)	Falsos Positivos (FP)
Predicción Negativa (N')	Falsos Negativos (FN)	Verdaderos Negativos (VN)
Verdaderos Positivos (VP): Son los casos en los que el modelo predijo correctamente la clase positiva.
Falsos Positivos (FP): Son los casos en los que el modelo predijo incorrectamente la clase positiva (predijo positivo cuando en realidad era negativo).
Falsos Negativos (FN): Son los casos en los que el modelo predijo incorrectamente la clase negativa (predijo negativo cuando en realidad era positivo).
Verdaderos Negativos (VN): Son los casos en los que el modelo predijo correctamente la clase negativa.
Funcionamiento y Uso
Evaluación del Rendimiento del Modelo: La matriz de confusión permite calcular diversas métricas de evaluación del rendimiento del modelo, como precisión, sensibilidad (recall), especificidad y F1-score, entre otras.

Comparación de Modelos: Permite comparar diferentes modelos de clasificación para determinar cuál tiene un mejor desempeño en términos de su capacidad para predecir correctamente las clases.

Ajuste de Umbrales de Decisión: Ayuda a ajustar los umbrales de decisión del modelo. Por ejemplo, en problemas de clasificación binaria, si el modelo produce muchas predicciones falsas positivas, puede ser útil ajustar el umbral para mejorar la precisión.

Identificación de Problemas de Clasificación: Facilita la identificación de los tipos de errores que comete el modelo. Por ejemplo, si hay muchos falsos negativos, podría indicar que el modelo necesita mejorar su capacidad para detectar casos positivos.

Ejemplo
Supongamos que tenemos un modelo de detección de spam por correo electrónico. Después de entrenarlo y evaluarlo con un conjunto de datos de prueba, obtenemos la siguiente matriz de confusión:

Spam Real (P)	No Spam Real (N)
Predicción Spam (P')	95 (VP)	5 (FP)
Predicción No Spam (N')	10 (FN)	1890 (VN)
En este ejemplo:

Hay 95 correos que realmente son spam (clase positiva) y el modelo los identificó correctamente (verdaderos positivos).
Hay 10 correos que realmente son spam pero el modelo los clasificó como no spam (falsos negativos).
Hay 5 correos que no son spam pero el modelo los clasificó incorrectamente como spam (falsos positivos).
Hay 1890 correos que no son spam y el modelo los clasificó correctamente como no spam (verdaderos negativos).
Conclusión
La matriz de confusión es una herramienta esencial para entender cómo se comporta un modelo de clasificación en términos de errores y aciertos. Es útil tanto para evaluar la precisión global del modelo como para identificar áreas específicas donde el modelo puede necesitar mejoras o ajustes.

Ejemplo de Regresión (Predicción de Calidad Educativa)

````python
# Importar librerías necesarias
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Crear listas de valores para cada país y sus resultados educativos
country_names = ['USA', 'Canada', 'Mexico']
math_scores = [85, 78, 70]  # Puntajes en matemáticas
reading_scores = [82, 80, 75]  # Puntajes en lectura
education_quality = ['Alta', 'Media', 'Baja']  # Calidad de la educación

# Crear un diccionario con los datos
data_dict = {
    "Country": country_names,
    "Math Scores": math_scores,
    "Reading Scores": reading_scores,
    "Education Quality": education_quality
}

# Crear un DataFrame a partir del diccionario
df_education = pd.DataFrame(data_dict)

# Aplicar One-Hot Encoding a la columna 'Education Quality'
# Primero, creamos el codificador OneHotEncoder
encoder = OneHotEncoder(sparse=False)

# Luego, ajustamos y transformamos la columna 'Education Quality'
education_encoded = encoder.fit_transform(df_education[['Education Quality']])

# Convertimos la salida codificada en un DataFrame y añadimos las columnas
education_encoded_df = pd.DataFrame(education_encoded, columns=encoder.categories_[0])

# Concatenamos el DataFrame original con el nuevo DataFrame codificado
df_education_encoded = pd.concat([df_education, education_encoded_df], axis=1)

# Eliminamos la columna original 'Education Quality'
df_education_encoded.drop(columns=['Education Quality'], inplace=True)

# Imprimir el DataFrame codificado
print(df_education_encoded)
````

````python
from sklearn.linear_model import LinearRegression

# Variables independientes (X) y variable dependiente (y)
X = df_education_encoded[['Math Scores', 'Reading Scores', 'Alta', 'Media', 'Baja']]  # Se incluyen las columnas codificadas
y = df_education['Math Scores']  # Por ejemplo, se podría predecir los puntajes en matemáticas

# Entrenar el modelo de regresión lineal
model = LinearRegression()
model.fit(X, y)

# Predecir los puntajes en matemáticas para un nuevo país con puntajes [80, 78] y calidad 'Alta'
new_data = [[80, 78, 1, 0, 0]]  # 'Alta' está representado como [1, 0, 0]
predicted_scores = model.predict(new_data)

print("Predicted Math Scores:", predicted_scores)
````
