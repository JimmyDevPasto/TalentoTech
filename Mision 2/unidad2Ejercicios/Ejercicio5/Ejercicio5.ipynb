```python
import keras
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd # Import the pandas library

#cargar el conjunto de datos MNIST
mnist = keras.datasets.mnist
#cargue la división de entrenamiento y prueba del conjunto de datos mnist
(training_images, training_labels), (test_images,test_labels) = mnist.load_data()

#puedes poner aqui entre 0 y 5999
index=1

#Imprime la etiqueta y la imagen.
np.set_printoptions(linewidth=320);
print(f'Label:{training_labels[index]}')
print(f'Image:\n{training_images[index]}')

#Muestra la imagen
plt.imshow(training_images[index])

#normalizarmos los valores de pizeles del tren y pobar las imagense
training_images = training_images/255.0
test_images = test_images/255.0
#construstruir el modelo de clasificación.

model = keras.Sequential([keras.layers.Flatten(),
                          keras.layers.Dense(128, activation='relu'),
                          keras.layers.Dense(10, activation='softmax')])

#compilacion del modelo
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

#entrenar el modelo
history=model.fit(training_images, training_labels, epochs=10)
#graficar el hostiral de entrenamiento
pd.DataFrame(history.history).plot(grid=True)

model.summary()
#Evaluar el modelo en el conjuto de entrenamiento
print('//// evalucion del modelo del conjunto de entrenamiento')
loss, accuracy = model.evaluate(training_images, training_labels)
print('Pérdida en el conjunto de entrenamiento :',loss)
print(f'Precisión en el conjunto de entrenamiento:{accuracy}')
print('//// fin evalucion del modelo del conjunto de entrenamiento \n')

print('//// evalucion del modelo del conjunto de datos no vistos')
#evaluar el modelo con datos no vistos
loss,accuracy = model.evaluate(test_images, test_labels)
print('Pérdida en el conjunto de entrenamiento :',loss)
print(f'Precisión en el conjunto de entrenamiento:{accuracy}')
print('////fin evalucion del modelo del conjunto de datos no vistos \n')

print('////evalucion del modelo del conjunto de datos no vistos predict')


#Evaluar el modelo con datos no vistos
# loss,accuracy = model.predict(test_images, test_labels)
# print('Pérdida en el conjunto de  prueba:',loss)
# print(f'Precisión en el conjunto de prueba:{accuracy}')
# print('////fin evalucion del modelo del conjunto de datos no vistos predict \n')

#sugerencia  gimini

# #Evaluar el modelo con datos no vistos
# predictions = model.predict(test_images)  # Get predictions instead of loss and accuracy
# # You can further process 'predictions' to obtain metrics like accuracy if needed.
# # For example, to get predicted classes:
# predicted_classes = np.argmax(predictions, axis=1)
# print('Predicted classes:', predicted_classes)
# print('////\n fin evalucion del modelo del conjunto de datos no vistos predict')

#Tarea
 #1 variar el numero de capas
 #1 variar el numero de epocas
 #3 variar mas las metricas de desempeño:
#Regresion (RMSE , R2)
# clasificacion matriz de confusion, f1
#para ambas graficar en un histograma la distribucion del error.
#error = y_test - y_pred
```

```python
import keras
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score, confusion_matrix, f1_score, classification_report

# Cargar el conjunto de datos MNIST
mnist = keras.datasets.mnist
(training_images, training_labels), (test_images, test_labels) = mnist.load_data()

#volver dataframes analisis exploratorio de los datos () tarea
# Puedes poner aquí entre 0 y 5999
index = 1

# Imprime la etiqueta y la imagen
np.set_printoptions(linewidth=320)
print(f'Label: {training_labels[index]}')
print(f'Image:\n{training_images[index]}')

# Muestra la imagen
plt.imshow(training_images[index], cmap='gray')
plt.show()

# Normalizamos los valores de píxeles del tren y prueba de las imágenes
training_images = training_images / 255.0
test_images = test_images / 255.0

# Construir el modelo de clasificación
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

# Compilación del modelo
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Entrenar el modelo
history = model.fit(training_images, training_labels, epochs=10)

# Graficar el historial de entrenamiento
pd.DataFrame(history.history).plot(grid=True)
plt.show()

# Resumen del modelo
model.summary()

# Evaluar el modelo en el conjunto de entrenamiento
print('//// Evaluación del modelo en el conjunto de entrenamiento')
loss, accuracy = model.evaluate(training_images, training_labels)
print('Pérdida en el conjunto de entrenamiento:', loss)
print(f'Precisión en el conjunto de entrenamiento: {accuracy}')
print('//// Fin evaluación del modelo en el conjunto de entrenamiento\n')

# Evaluar el modelo en el conjunto de datos no vistos
print('//// Evaluación del modelo en el conjunto de datos no vistos')
loss, accuracy = model.evaluate(test_images, test_labels)
print('Pérdida en el conjunto de prueba:', loss)
print(f'Precisión en el conjunto de prueba: {accuracy}')
print('//// Fin evaluación del modelo en el conjunto de datos no vistos\n')

# Predicciones con datos no vistos
predictions = model.predict(test_images)
predicted_labels = np.argmax(predictions, axis=1)

# Calcular RMSE y R2
rmse = np.sqrt(mean_squared_error(test_labels, predicted_labels))
r2 = r2_score(test_labels, predicted_labels)
print(f'RMSE: {rmse}')
print(f'R2: {r2}')

# Calcular matriz de confusión y F1-score
conf_matrix = confusion_matrix(test_labels, predicted_labels)
f1 = f1_score(test_labels, predicted_labels, average='weighted')
print(f'Matriz de Confusión:\n{conf_matrix}')
print(f'F1-score: {f1}')
print(classification_report(test_labels, predicted_labels))

# Graficar histogramas de la distribución del error
errors = test_labels - predicted_labels

plt.figure(figsize=(10, 5))
plt.hist(errors, bins=100, edgecolor='k')
plt.xlabel('Error')
plt.ylabel('Frecuencia')
plt.title('Distribución del Error')
plt.grid(True)
plt.show()
```
